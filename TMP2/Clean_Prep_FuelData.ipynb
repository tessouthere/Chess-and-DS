```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('alt_fuel_stations.csv')
```

```python
df.head()
```

```python
def get_region(state, region):
    if state in region['WEST']:
        return 'WEST'
    elif state in region['SOUTHEAST']:
        return 'SOUTHEAST'
    elif state in region['MIDWEST']:
        return 'MIDWEST'
    elif state in region['SOUTHWEST']:
        return 'SOUTHWEST'
    elif state in region['NORTHEAST']:
        return 'NORTHEAST'
    else:
        return None

WEST = ['WA', "OR", "ID", "MT","WY","NV","CA","UT","CO","AK","HI"]
SOUTHEAST =  ["MD","DE","DC","WV",'VA',"KY","TN","NC","SC","GA",'FL',"AL","MS","LA","AR","PR"]
MIDWEST = ["ND","MN","WI","MI","SD","IA","NE","IL","IN","OH","MO","KS"]
SOUTHWEST = ['AZ', 'NM','OK', 'TX']
NORTHEAST = ["ME","VT","NH","MA","RI","CT","NJ","PA","NY"]

region = {'WEST': WEST, 'SOUTHEAST': SOUTHEAST, 'MIDWEST': MIDWEST, 'SOUTHWEST': SOUTHWEST, 'NORTHEAST': NORTHEAST}
df['Region'] = df['State'].apply(get_region, args=(region,))
df.head()
```

```python
df.to_csv('fuel_stations.csv', index=False)
```
